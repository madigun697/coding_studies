```python
import pmat_baker_bird
import pandas as pd
from IPython.display import display
```

```python
t = ['bbabbab', 'aacacba', 'bbbacac', 'acabbab', 'caacaba', 'bbbbacc', 'accabab']
display(pd.DataFrame([list(r) for r in t]))

print('--------------------------------------------')

p = ['acac', 'bbab', 'caba']

print('Pattern is')
display(pd.DataFrame([list(r) for r in p]))
result = pmat_baker_bird.patternSearch(t, p)
print(result)

for r in result:
    display(pd.DataFrame([list(r) for r in t]).iloc[r[0] - len(p) + 1:r[0]+1, r[1] - len(p[0]) + 1:r[1]+1])

print('--------------------------------------------')

p = ['aca', 'bba', 'cab', 'bac']

print('Pattern is')
display(pd.DataFrame([list(r) for r in p]))
result = pmat_baker_bird.patternSearch(t, p)
print(result)

for r in result:
    display(pd.DataFrame([list(r) for r in t]).iloc[r[0] - len(p) + 1:r[0]+1, r[1] - len(p[0]) + 1:r[1]+1])

print('--------------------------------------------')

p = ['aca', 'bba', 'cab']

print('Pattern is')
display(pd.DataFrame([list(r) for r in p]))
result = pmat_baker_bird.patternSearch(t, p)
print(result)

for r in result:
    display(pd.DataFrame([list(r) for r in t]).iloc[r[0] - len(p) + 1:r[0]+1, r[1] - len(p[0]) + 1:r[1]+1])
```
